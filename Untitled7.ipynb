Построим матрицу корреляций

```python
corr = personal_loan.corr()
corr.style.background_gradient(cmap='coolwarm')
```

Мы видим, что опыт и возраст имеют высокую положительную корреляцию (это также видно на графике выше).

Что касается таргета, то есть personal loan, то данный атрибут больше коррелирует с атрибатами income, CCAvg, CD Account.

Проверим, насколько данные корреляции достоверные

```python
headers = ['Age', 'Experience', 'Income', 'CCAvg', 'Personal Loan', 'CD Account']
```

```python
def correl(headers: list) -> list:
  pairs = []
  correlate = []
  for ind in range(len(headers)):
    for i in range(len(headers)):
      try:
         pair = (headers[ind], headers[ind + i])
         if pair[0] != pair[1]:
           pairs.append(pair)
      except IndexError:
         pass
  for pair in pairs:
     rho, p = spearmanr(personal_loan[pair[0]], personal_loan[pair[1]])
     correlate.append([pair, rho, p])
  return correlate
```

```python
corr_list = correl(headers)
my = pd.DataFrame(corr_list)
my.columns = ['pairs', 'корреляция Спирмена', 'значение p для интерпретации значимости коэффициента']
my
```

Если р < 0.05, то корреляция достоверная

```python
for idx, number in enumerate(my['значение p для интерпретации значимости коэффициента']):
  if number > 0.05:
    my.drop(idx, inplace=True)
  
my
```

Мы видим, что возраст и опыт - коррелированные признаки, поэтому один из них необходимо удалить

Напишем функцию, которая будет выводить результат в более наглядный вид

```python
def results(results):
  print(f'Optimal Hyperparams: {results.best_params_}')
  means = results.cv_results_['mean_test_score']
  stds = results.cv_results_['std_test_score']

  for mean, std, params in zip(means, stds, results.cv_results_['params']):
    print(f'Mean {round(mean, 3)}, Standart Deviation {round(std * 2, 3)} Hyperparameters {params}')
```

```python
results(cross_val)
```

# Полносвязная нейронная сеть

# Модель сети

```python
def create_network():
    Net = Sequential()
    Net.add(Dense(64, activation = "relu"))
    Net.add(Dense(1, activation='sigmoid'))
    Net.compile(
        loss = "binary_crossentropy",
        optimizer = "adam",
        metrics=[keras.metrics.BinaryAccuracy(),'accuracy']
    )
    return Net
```

# Cross Validation of neural network

```python
neural_network = KerasClassifier(build_fn=create_network, epochs=5, batch_size=20,
                                     verbose=0)

scores = cross_val_score(neural_network, X_train, y_train.values.ravel() ,cv=5, scoring = 'accuracy' )
print(scores.mean())
```

# Accuracy, recall, precision of neural network

```python
Network = create_network()
hist = Network.fit(X_train, y_train, epochs=5,validation_data=(X_test, Y_test), batch_size=40)


acc = hist.history['accuracy']
val = hist.history['val_accuracy']
epochs = range(1, len(acc) + 1)
 
plt.plot(epochs, acc, '-', label='Training accuracy')
plt.plot(epochs, val, ':', label='Validation accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.plot()
print()
print('--------Predict--------')
network_predict = Network.predict(X_test) >= 0.5
print("Да:", sum(network_predict==1))
print("Нет:", sum(network_predict==0))
```

```python
accuracy = round(accuracy_score(Y_test, network_predict), 3)
precision = round(precision_score(Y_test, network_predict), 3)
recall = round(recall_score(Y_test, network_predict), 3)

print(f'Accuracy: {accuracy}, Precision: {precision}, Recall: {recall} of neural network on test set')
```

# Confusion Matrix of neural network
